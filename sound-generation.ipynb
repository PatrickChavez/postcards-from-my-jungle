```python
import torch
import soundfile as sf
from diffusers import StableAudioPipeline
from IPython.display import Audio  # Import the Audio module for playback
from diffusers import AudioLDMPipeline
import scipy

repo_id = "cvssp/audioldm-s-full-v2"
# Removed torch_dtype=torch.float16 for CPU compatibility
pipe = AudioLDMPipeline.from_pretrained(repo_id)
pipe = pipe.to("cpu")

prompt = "The great frog-shark does a mighty roar."
audio = pipe(prompt, num_inference_steps=50, audio_length_in_s=5.0).audios[0]

scipy.io.wavfile.write("animalSound.wav", rate=16000, data=audio)

Audio(audio, rate=16000)
```
